# Shortcuts
There are many shortcuts for Jupyter notebook.<br> 
In the tool bar above click "Help"-> "Keyboard Shortcuts" <br> 
You can also google "Jupyter notebook shortcuts". Here are some that I find myself using often:

* In edit mode (green cell)
    - Alt + Enter = Runs cell and creates new cell below
    - Shift + Enter = Runs cell and goes to the next one. Creates new one if there is not one
    - ESC = Leaves edit mode

* In command mode (blue cell)
    - c = Copy selected cell
    - v = Pastes copied cell below
    - DD (hit twice back-to-back) = Deletes cell
